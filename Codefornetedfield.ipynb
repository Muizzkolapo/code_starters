`````python
def build_type_definition(field):
    """Builds the data type definition for a field."""
    field_type = field['type']
    field_mode = field.get('mode', '')

    if field_type == "RECORD":
        fields_def = ', '.join([f"{f['name']} {build_type_definition(f)}" for f in field['fields']])
        record_type = f"STRUCT<{fields_def}>"
        return f"ARRAY<{record_type}>" if field_mode == "REPEATED" else record_type
    elif field_type == "INTEGER":
        return "ARRAY<INT64>" if field_mode == "REPEATED" else "INT64"
    elif field_type == "STRING":
        return "ARRAY<STRING>" if field_mode == "REPEATED" else "STRING"
    elif field_type == "NUMERIC":
        return "ARRAY<NUMERIC>" if field_mode == "REPEATED" else "NUMERIC"
    elif field_type == "BOOLEAN":
        return "ARRAY<BOOLEAN>" if field_mode == "REPEATED" else "BOOLEAN"
    return field_type

def cast_null_to_type(value_type):
    """Cast null values to the appropriate SQL type."""
    if value_type == "STRING":
        return "CAST(NULL AS STRING)"
    elif value_type == "INTEGER":
        return "CAST(NULL AS INT64)"
    elif value_type == "NUMERIC":
        return "CAST(NULL AS NUMERIC)"
    elif value_type == "BOOLEAN":
        return "CAST(NULL AS BOOLEAN)"
    return "NULL"

def format_sql_value(value, value_type):
    """Format value based on its SQL type."""
    if value is None:
        return cast_null_to_type(value_type)

    if value_type == "STRING":
        if "'" in value and '"' not in value:
            return f'"{value}"'  # Use double quotes
        else:
            return "'" + value.replace("'", "''") + "'"  # Use single quotes and escape single quotes
    elif value_type == "TIMESTAMP" or value_type == "DATE":
        return f"'{value}'"  # Timestamps and dates should be quoted
    elif value_type in ["INTEGER", "NUMERIC"]:
        return str(value)
    elif value_type == "BOOLEAN":
        return "TRUE" if value else "FALSE"
    return str(value)

def build_sql_value(data, field):
    """Build SQL value based on field schema."""
    field_type = field['type']
    field_mode = field.get('mode', '')

    if field_type == "RECORD":
        if field_mode == "REPEATED":
            return '[' + ', '.join(['(' + build_sql_struct(item, field['fields']) + ')' for item in data]) + ']'
        else:
            return '(' + build_sql_struct(data, field['fields']) + ')'
    else:
        return format_sql_value(data, field_type)

def build_sql_struct(data, schema):
    """Build SQL struct based on schema."""
    struct_values = []
    for field in schema:
        field_name = field['name']
        field_data = data.get(field_name, None)
        sql_value = build_sql_value(field_data, field)
        struct_values.append(sql_value)
    return ', '.join(struct_values)

def build_sql_query(data, schema):
    """Build SQL query based on structured data and schema."""
    query_parts = []
    for field in schema:
        field_name = field['name']
        field_data = data.get(field_name, None)
        field_type_def = build_type_definition(field)
        sql_value = build_sql_value(field_data, field)

        # Include data type definition only for complex (RECORD or REPEATED) fields
        if field['type'] == 'RECORD' or field.get('mode', '') == 'REPEATED':
            query_part = f"{field_type_def} {sql_value} AS {field_name}"
        else:
            query_part = f"{sql_value} AS {field_name}"

        query_parts.append(query_part)
    return "SELECT\n  " + ',\n  '.join(query_parts)
`````

`````python
# Wporking code
# Wporking code
def build_type_definition(field):
    """Builds the data type definition for a field."""
    field_type = field['type']
    field_mode = field.get('mode', '')

    if field_type == "RECORD":
        fields_def = ', '.join([f"{f['name']} {build_type_definition(f)}" for f in field['fields']])
        record_type = f"STRUCT<{fields_def}>"
        return f"ARRAY<{record_type}>" if field_mode == "REPEATED" else record_type
    elif field_type == "INTEGER":
        return "ARRAY<INT64>" if field_mode == "REPEATED" else "INT64"
    elif field_type == "STRING":
        return "ARRAY<STRING>" if field_mode == "REPEATED" else "STRING"
    elif field_type == "NUMERIC":
        return "ARRAY<NUMERIC>" if field_mode == "REPEATED" else "NUMERIC"
    elif field_type == "BOOLEAN":
        return "ARRAY<BOOLEAN>" if field_mode == "REPEATED" else "BOOLEAN"
    return field_type

def cast_null_to_type(value_type):
    """Cast null values to the appropriate SQL type."""
    if value_type == "STRING":
        return "CAST(NULL AS STRING)"
    elif value_type == "INTEGER":
        return "CAST(NULL AS INT64)"
    elif value_type == "NUMERIC":
        return "CAST(NULL AS NUMERIC)"
    elif value_type == "BOOLEAN":
        return "CAST(NULL AS BOOLEAN)"
    return "NULL"

def format_sql_value(value, value_type):
    """Format value based on its SQL type."""
    if value is None:
        return cast_null_to_type(value_type)

    if value_type == "STRING":
        if "'" in value and '"' not in value:
            return f'"{value}"'  # Use double quotes
        else:
            return "'" + value.replace("'", "''") + "'"  # Use single quotes and escape single quotes
    elif value_type == "TIMESTAMP" or value_type == "DATE":
        return f"'{value}'"  # Timestamps and dates should be quoted
    elif value_type in ["INTEGER", "NUMERIC"]:
        return str(value)
    elif value_type == "BOOLEAN":
        return "TRUE" if value else "FALSE"
    return str(value)

def build_sql_value(data, field):
    """Build SQL value based on field schema."""
    field_type = field['type']
    field_mode = field.get('mode', '')

    if field_type == "RECORD":
        if field_mode == "REPEATED":
            return '[' + ', '.join(['(' + build_sql_struct(item, field['fields']) + ')' for item in data]) + ']'
        else:
            return '(' + build_sql_struct(data, field['fields']) + ')'
    else:
        return format_sql_value(data, field_type)

def build_sql_struct(data, schema):
    """Build SQL struct based on schema."""
    struct_values = []
    for field in schema:
        field_name = field['name']
        field_data = data.get(field_name, None)
        sql_value = build_sql_value(field_data, field)
        struct_values.append(sql_value)
    return ', '.join(struct_values)

def build_sql_query(data, schema):
    """Build SQL query based on structured data and schema."""
    query_parts = []
    for field in schema:
        field_name = field['name']
        field_data = data.get(field_name, None)
        field_type_def = build_type_definition(field)
        sql_value = build_sql_value(field_data, field)

        # Include data type definition only for complex (RECORD or REPEATED) fields
        if field['type'] == 'RECORD' or field.get('mode', '') == 'REPEATED':
            query_part = f"{field_type_def} {sql_value} AS {field_name}"
        else:
            query_part = f"{sql_value} AS {field_name}"

        query_parts.append(query_part)
    return "SELECT\n  " + ',\n  '.join(query_parts)

def build_union_all_sql_queries(data_list, schema):
    """Build SQL queries for a list of data items and combine them with UNION ALL."""
    queries = [build_sql_query(data_item, schema) for data_item in data_list]
    return '\nUNION ALL\n'.join(queries)



# Generate SQL queries for all items in the list and combine with UNION ALL
combined_sql_query = build_union_all_sql_queries(data_list, schema)
print(combined_sql_query)
`````
